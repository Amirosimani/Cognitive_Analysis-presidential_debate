```python
import json
import csv
```

```python
"""
This can probably be improved but serves the purpose of our analysis.
I used Chris's schema so it will be useful for Quarz analysis.

"""

EMOTIONS = [
    'contempt',
    'fear',
    'sadness',
    'disgust',
    'anger',
    'neutral',
    'happiness',
    'surprise',
]


CSV_COLUMNS = [
    'url',
    'frame_time',
    'trump_happiness',
    'trump_neutral',
    'trump_anger',
    'trump_contempt',
    'trump_suprise',
    'trump_sadness',
    'trump_fear',
    'trump_disgust',
]
```

```python
with open('/Users/Amiros/GitHub/debate/acceptance2.json') as data_file:    
    json_data = json.load(data_file)
    
json_data = {int(k):v for k,v in json_data.items()} #converting keys to integer   
frame_keys = sorted(json_data.keys())
```

```python
frame_keys
```

```python
csv_data = []
   
for key in frame_keys:
    url = 'https://s3-us-west-2.amazonaws.com/post-election/inauguration/inaugruation%d.jpg' % key
    frame = json_data[key]
        
    row = {
        #'frame': frame,
        'url': url,
        'frame_time': key,
    }
    
    if len(frame) > 2:
        print('Frame %s has more than 2 faces (%s)' % (key, url))
        continue
    if len(frame) > 1:
        for face in frame:
            if face['faceAttributes']['gender'] == 'male' and face['faceRectangle']['left'] > 500 and face['faceRectangle']['left'] < 750:
                prefix = 'trump_'
                for emotion in EMOTIONS:
                    #print(face['scores'][emotion])
                    row[prefix + emotion] = face['scores'][emotion]
                ### assign emotions vale to each candidate and change the name to the respective one 
                    
            row.pop("frame", None)        
      
    elif len(frame) == 2:
        for face in frame:
            if face['faceAttributes']['gender'] == 'male':
                prefix = 'trump_'
                try:
                    for emotion in EMOTIONS:
                    #print(face['scores'][emotion])
                        row[prefix + emotion] = face['scores'][emotion]
                except:
                    pass
                ### assign emotions vale to each candidate and change the name to the respective one 
                    
        row.pop("frame", None)        

    csv_data.append(row)
```

```python
import pandas as pd
df = pd.DataFrame(csv_data)
```
